```python
data.rename(columns={'Gender Groups Y/N':'Gender_groups', 'Race Groups Y/N':'Race_groups','Gender Comps Y/N':'Gender_comps','Race Comps Y/N':'Race_comps','Empowerment Sent':'Empowerment_Sent'}, inplace=True)
```

```python
data.rename(columns={'Gender Group Sent':'Gender_Group_Sent', 'Race Group Sent':'Race_Group_Sent','Gender Comp Sent':'Gender_Comp_Sent','Race Comp Sent':'Race_Comp_Sent','Improvement Sent':'Improvement_Sent'}, inplace=True)
```

```python
def transform(x): #pie chart
  if x == "negative":
    return 'negative'
  elif x == "positive":
    return 'positive'
  else:
    return 'neutral'

groups_of_empow = data.Empowerment_Sent.apply(transform)

values = groups_of_empow.value_counts()
labels = values.index
colors = ['brown', 'blue', 'green']

fig = plt.figure(figsize=(20, 10))
plt.pie(values, colors=colors, autopct='%1.1f%%', startangle=90, textprops={ 'color': 'w' })

plt.title('Empowerment Sentiment', fontsize=15)
plt.legend(labels)
plt.show()
```

```python
def transform(x): #pie chart
  if x == "negative":
    return 'negative'
  elif x == "positive":
    return 'positive'
  else:
    return 'neutral'

groups_of_impro = data.Improvement_Sent.apply(transform)

values = groups_of_impro.value_counts()
labels = values.index
colors = ['red', 'blue', 'pink']

fig = plt.figure(figsize=(20, 10))
plt.pie(values, colors=colors, autopct='%1.1f%%', startangle=90, textprops={ 'color': 'w' })

plt.title('Improvement Sentiment', fontsize=15)
plt.legend(labels)
plt.show()
```
